```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Define file paths for four songs from different genres (Replace with actual file paths)
audio_files = {
    "Classical -Jab Dil Hi Toot gaya Hum Ji Ke Kya": "/content/drive/MyDrive/songs/Clasical_song.mp3",
    "Rock - Ye Parda Hata Do": "/content/drive/MyDrive/songs/Rock.mp3",
    "Hip-Hop - Lagav e Lu Jab Lipistick": "/content/drive/MyDrive/songs/Hip-Hop.mp3",
    "Electronic - One More Time": "/content/drive/MyDrive/songs/Electronic_song.mp3"
}

def plot_spectrogram(audio_path, title):
    y, sr = librosa.load(audio_path, sr=None)  # Load audio file

    # Compute Short-Time Fourier Transform (STFT)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

    # Plot spectrogram
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.xlabel("Time (s)")
    plt.ylabel("Frequency (Hz)")
    plt.show()

# Generate spectrograms for all songs
for title, file in audio_files.items():
    plot_spectrogram(file, title)
```
